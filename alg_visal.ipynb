```python
df=pd.read_sql_query(query,os_rdb_engine)
```

```python
col=list(df.columns)
```

```python
col=[x for x in col if x[0:2]=='CO' or x[0:2]=='RE' or x[0:2]=='AD' or x[0:2]=='RO'] 
```

```python
col.remove(col[-1])
```

```python
col_co=[x for x in col if x[0:2]=='CO']
col_re=[x for x in col if x[0:2]=='RE']
col_ad=[x for x in col if x[0:2]=='AD']
col_ro=[x for x in col if x[0:2]=='RO']
```

```python
res_col=[col_co,col_re,col_ad,col_ro]
```

```python
f1 = plt.figure(figsize=(30,20))
for idx in res_col:
    for k in idx:
        if k[0:2]=='CO':
            x1=[ x for x in df[k].tolist() if str(x) != 'nan']
            plt.subplot(221)
            sns.distplot(x1, hist = False, label = '%s'%(k),
                kde_kws = {'shade': True, 'linewidth': 3})
            plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
            plt.axvline(x=1, color='black', linestyle='--', linewidth=1)
            plt.title('CO_NAME')
        if k[0:2]=='RE':
            x2=[ x for x in df[k].tolist() if str(x) != 'nan']
            plt.subplot(222)
            sns.distplot(x2, hist = False, label = '%s'%(k),
                kde_kws = {'shade': True, 'linewidth': 3})
            plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
            plt.axvline(x=1, color='black', linestyle='--', linewidth=1)
            plt.title('REP_PHONE')
        if k[0:2]=='AD':
            x3=[ x for x in df[k].tolist() if str(x) != 'nan']
            plt.subplot(223)
            sns.distplot(x3, hist = False, label = '%s'%(k),
                kde_kws = {'shade': True, 'linewidth': 3})
            plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
            plt.axvline(x=1, color='black', linestyle='--', linewidth=1)
            plt.title('ADDR')
        if k[0:2]=='RO':
            x4=[ x for x in df[k].tolist() if str(x) != 'nan']
            plt.subplot(224)
            sns.distplot(x4, hist = False, label = '%s'%(k),
                kde_kws = {'shade': True, 'linewidth': 3})
            plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
            plt.axvline(x=1, color='black', linestyle='--', linewidth=1)
            plt.title('REP_PHONE')
plt.show()
f1.savefig('4features_by_num.png')
```
